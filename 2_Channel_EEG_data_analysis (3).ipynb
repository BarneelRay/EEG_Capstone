We now filter the data using bandpass filters to remove noise

```python
import pandas as pd
from scipy.signal import butter, filtfilt

def bandpass_filter(data, lowcut, highcut, fs, order=5):
    """
    Applies a bandpass Butterworth filter to the input data.

    Args:
        data (pd.DataFrame or pd.Series): The input data.
        lowcut (float): The lower cutoff frequency in Hz.
        highcut (float): The upper cutoff frequency in Hz.
        fs (float): The sampling frequency in Hz.
        order (int, optional): The order of the filter. Defaults to 5.

    Returns:
        pd.DataFrame or pd.Series: The filtered data.
    """
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    filtered_data = filtfilt(b, a, data, axis=0)
    return filtered_data

# Example usage: Assuming 'av_ch1_funny' is your DataFrame and you want to filter all columns
# with a bandpass from 0.1 Hz to 30 Hz, and your sampling frequency is 128 Hz:

fs = 128  # assuming a 128 Hz sampling frequency
lowcut = 0.3
highcut = 40

# Apply the filter to all columns of the DataFrame
av_ch1_funny_filtered = av_ch1_funny.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
av_ch1_sad_filtered = av_ch1_sad.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
av_ch1_scary_filtered = av_ch1_scary.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
av_ch2_funny_filtered = av_ch2_funny.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
av_ch2_sad_filtered = av_ch2_sad.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
av_ch2_scary_filtered = av_ch2_scary.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
a_ch1_enjoyable_filtered = a_ch1_enjoyable.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
a_ch1_relaxed_filtered = a_ch1_relaxed.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
a_ch1_sad_filtered = a_ch1_sad.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
a_ch2_enjoyable_filtered = a_ch2_enjoyable.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
a_ch2_relaxed_filtered = a_ch2_relaxed.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
a_ch2_sad_filtered = a_ch2_sad.apply(lambda col: bandpass_filter(col, lowcut, highcut, fs), axis=0)
```

We implement PCA and convolution to merge the data pertaining to both channels and thus producing two new signals that will be used to generate datasets for both kind of stimuli and combination method.

```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np

def process_data(df1, df2):
    """
    Applies PCA and convolution to corresponding subject data from
    two dataframes and returns two dataframes with one column per subject.

    Args:
        df1: First dataframe.
        df2: Second dataframe.

    Returns:
        Two dataframes:
            - pca_df: DataFrame containing PCA results for each subject.
            - conv_df: DataFrame containing convolution results for each subject.
    """

    # Get subject column names from both DataFrames
    subjects = df1.columns  # Assuming both DataFrames have the same columns (subjects)

    # Initialize dictionaries to store PCA and convolution results
    pca_results = {}
    conv_results = {}

    # Iterate through each subject
    for subject in subjects:
        # Select data for the current subject from both DataFrames
        subject_data_df1 = df1[subject].values.reshape(-1, 1)
        subject_data_df2 = df2[subject].values.reshape(-1, 1)

        # Handle potential shape mismatch due to different numbers of subjects
        min_len = min(len(subject_data_df1), len(subject_data_df2))
        subject_data_df1 = subject_data_df1[:min_len]
        subject_data_df2 = subject_data_df2[:min_len]

        # Replace NaN values with 0 in subject data
        subject_data_df1 = np.nan_to_num(subject_data_df1, nan=0.0)
        subject_data_df2 = np.nan_to_num(subject_data_df2, nan=0.0)

        # Combine the subject data
        subject_data = np.concatenate([subject_data_df1, subject_data_df2], axis=1)

        # Apply PCA
        pca = PCA(n_components=1)  # Reduce to 1 component
        pca_result = pca.fit_transform(subject_data)
        pca_results[subject] = pca_result.flatten()  # Store in dictionary

        # Apply Convolution and take the mean
        convolution_result = np.convolve(subject_data_df1.flatten(), subject_data_df2.flatten(), mode='full')
        conv_results[subject] = convolution_result  # Store in dictionary

    # Create DataFrames for PCA and Convolution results
    pca_df = pd.DataFrame(pca_results)
    conv_df = pd.DataFrame(conv_results)
    print(f"combined {df1} and {df2}.")
    return pca_df, conv_df
av_funny_pca, av_funny_conv = process_data(av_ch1_funny_filtered, av_ch2_funny_filtered)
av_sad_pca, av_sad_conv = process_data(av_ch1_sad_filtered, av_ch2_sad_filtered)
av_scary_pca, av_scary_conv = process_data(av_ch1_scary_filtered, av_ch2_scary_filtered)
a_enjoyable_pca, a_enjoyable_conv = process_data(a_ch1_enjoyable_filtered, a_ch2_enjoyable_filtered)
a_relaxed_pca, a_relaxed_conv = process_data(a_ch1_relaxed_filtered, a_ch2_relaxed_filtered)
a_sad_pca, a_sad_conv = process_data(a_ch1_sad_filtered, a_ch2_sad_filtered)
```

We merge dataframes to generate new data.

```python
def process_data(df1, df2):
    """
    Applies PCA and convolution to corresponding subject data from
    two dataframes and returns two dataframes with one column per subject.

    Args:
        df1: First dataframe.
        df2: Second dataframe.

    Returns:
        Two dataframes:
            - pca_df: DataFrame containing PCA results for each subject.
            - conv_df: DataFrame containing convolution results for each subject.
    """

    # Get subject column names from both DataFrames
    subjects = df1.columns  # Assuming both DataFrames have the same columns (subjects)

    # Initialize dictionaries to store PCA and convolution results
    pca_results = {}
    conv_results = {}

    # Iterate through each subject
    for subject in subjects:
        # Check if the entire column for the subject is NaN in both DataFrames
        if df1[subject].isna().all() and df2[subject].isna().all():
            print(f"Skipping subject {subject} as the entire column is NaN in both DataFrames.")
            continue

        # Select data for the current subject from both DataFrames
        subject_data_df1 = df1[subject].values.reshape(-1, 1)
        subject_data_df2 = df2[subject].values.reshape(-1, 1)

        # Handle potential shape mismatch due to different numbers of subjects
        min_len = min(len(subject_data_df1), len(subject_data_df2))
        subject_data_df1 = subject_data_df1[:min_len]
        subject_data_df2 = subject_data_df2[:min_len]

        # Replace NaN values with 0 in subject data
        subject_data_df1 = np.nan_to_num(subject_data_df1, nan=0.0)
        subject_data_df2 = np.nan_to_num(subject_data_df2, nan=0.0)

        # Combine the subject data
        subject_data = np.concatenate([subject_data_df1, subject_data_df2], axis=1)

        # Skip subjects with zero variance or insufficient data
        if np.var(subject_data, axis=0).sum() == 0:
            print(f"Skipping PCA for subject {subject} due to zero variance.")
            continue
        if subject_data.shape[0] < 2:
            print(f"Skipping PCA for subject {subject} due to insufficient data.")
            continue

        # Apply PCA with error handling
        try:
            pca = PCA(n_components=1)  # Reduce to 1 component
            pca_result = pca.fit_transform(subject_data)
            pca_results[subject] = pca_result.flatten()  # Store in dictionary

            # Apply Convolution and take the mean
            convolution_result = np.convolve(subject_data_df1.flatten(), subject_data_df2.flatten(), mode='full')
            conv_results[subject] = convolution_result  # Store in dictionary

        except Exception as e:
            print(f"PCA failed for subject {subject}: {e}")
            continue

    # Create DataFrames for PCA and Convolution results
    pca_df = pd.DataFrame(pca_results)
    conv_df = pd.DataFrame(conv_results)
    return pca_df, conv_df
av_funny_pca, av_funny_conv = process_data(av_ch1_funny_filtered, av_ch2_funny_filtered)
av_sad_pca, av_sad_conv = process_data(av_ch1_sad_filtered, av_ch2_sad_filtered)
av_scary_pca, av_scary_conv = process_data(av_ch1_scary_filtered, av_ch2_scary_filtered)
a_enjoyable_pca, a_enjoyable_conv = process_data(a_ch1_enjoyable_filtered, a_ch2_enjoyable_filtered)
a_relaxed_pca, a_relaxed_conv = process_data(a_ch1_relaxed_filtered, a_ch2_relaxed_filtered)
a_sad_pca, a_sad_conv = process_data(a_ch1_sad_filtered, a_ch2_sad_filtered)
```

Preparing finally merged data after restructuring


```python
import pandas as pd
import numpy as np

def restructure_data(dataframes, emotion_labels):
    """
    Restructures and transposes dataframes to a new format with subject, data, and emotion columns.

    Args:
        dataframes: A list of pandas DataFrames to restructure.
        emotion_labels: A list of emotion labels corresponding to the DataFrames.

    Returns:
        A pandas DataFrame with columns: 'subject', 'data', and 'emotion'.
    """
    rows = []
    for df, emotion in zip(dataframes, emotion_labels):
        # Transpose the DataFrame so columns become rows
        transposed = df.T

        # Get original subject column names before transposing
        original_subjects = df.columns

        # Create rows with subject, data (as a NumPy array), and emotion
        for i, (subject, data) in enumerate(transposed.iterrows()):
            # Use original subject column names
            rows.append({'subject': original_subjects[i], 'data': data.to_numpy(), 'emotion': emotion})

    # Create a new DataFrame from the structured rows
    structured_df = pd.DataFrame(rows, columns=['subject', 'data', 'emotion'])
    return structured_df

# Restructure data for audio-visual (av) PCA
av_pca = restructure_data(
    [av_funny_pca, av_sad_pca, av_scary_pca],
    ['funny', 'sad', 'scary']
)

# Restructure data for audio-visual (av) Convolution
av_conv = restructure_data(
    [av_funny_conv, av_sad_conv, av_scary_conv],
    ['funny', 'sad', 'scary']
)

# Restructure data for audio (a) PCA
a_pca = restructure_data(
    [a_enjoyable_pca, a_relaxed_pca, a_sad_pca],
    ['enjoyable', 'relaxed', 'sad']
)

# Restructure data for audio (a) Convolution
a_conv = restructure_data(
    [a_enjoyable_conv, a_relaxed_conv, a_sad_conv],
    ['enjoyable', 'relaxed', 'sad']
)

print("av_pca:")
print(av_pca)
print("\nav_conv:")
print(av_conv)
print("\na_pca:")
print(a_pca)
print("\na_conv:")
print(a_conv)
```
